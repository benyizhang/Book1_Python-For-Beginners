````python
# Text preprocessing
# All text preprocess for column text_lower
import string
import re
import emoji
import zhon.hanzi
from unicodedata import normalize
import contractions

# Apply both cleaning steps
# remove http url and [\t\r\n\a]
df['Text_lower'] = df['Text_lower'].apply(lambda x: re.sub(r'[\t\r\n\a]', ' ', x))

# Remove any url link
df['Text_lower'] = df['Text_lower'].apply(lambda x: re.sub(r'(http\S+|html\S+)', '', x).strip())

# remove emojis, I hate it
df.loc[:,'Text_lower'] = df.loc[:,'Text_lower'].apply(lambda x: emoji.replace_emoji(x, replace=''))

# remove @, # and keep % replace it as percent
df.loc[:,'Text_lower'] = df.loc[:,'Text_lower'].apply(lambda x: x.replace("@", "").replace("#", "").replace("%", " percent"))

# normalise unicode text into strings, and keep it lower case
df.loc[:,'Text_lower'] = df.loc[:, 'Text_lower'].apply(lambda x: normalize('NFKD', x).lower())

# expand contraction words
# 've => have
# don't => do not etc
# change other words well.
contractions.add('co2','carbon emission')
contractions.add('ecb','european central bank')
contractions.add('ldcs','least developed countries')
contractions.add('sdg', 'sustainable development goals')
contractions.add('sdgs', 'sustainable development goals')
contractions.add('esg', 'environment, social, governance')
contractions.add('csr', 'corporate social responsibility')
df.loc[:,'Text_lower'] = df.loc[:, 'Text_lower'].apply(lambda x: contractions.fix(x))

# remove all English and Chinese punctuations
all_punctuation = ''.join([ch for ch in string.punctuation + zhon.hanzi.punctuation if ch not in '%'])
df.loc[:,'Text_lower'] = df.loc[:,'Text_lower'].apply(lambda x: x.translate(str.maketrans('', '', all_punctuation)))
del all_punctuation # don't need all_punctuation anymore

# remove unwanted symbols.
df.loc[:,'Text_lower'] = df.loc[:,'Text_lower'].apply(lambda x: re.sub(r'[^\w\s%]', '', x))

# remove unwanted white space
df.loc[:,'Text_lower'] = df.loc[:,'Text_lower'].apply(lambda x: re.sub(r'\s+', ' ', x).strip())

# keep only English letters
df['Text_lower'] = df['Text_lower'].apply(lambda x: re.sub(r'[^A-Za-z\s]', '', x))
df['Text_lower'] = df['Text_lower'].apply(lambda x: re.sub(r'\s+', ' ', x).strip())

# remove numebrs in my text. They do not provide any meaningful values
# df.loc[:, 'Text_lower'] = df.loc[:, 'Text_lower'].apply(lambda x: re.sub(r'\d+', '', x))
````

### Similarity test
The simiary test will be conducted based on working data. We can set up the GPU to make my over simialrity test to work faster.  

Eventhough we have removed the perfect duplicates, we still have text that have high level similarity. Given that I have approximately 217,600 pieces of text information, my similarity test will produce a 217,600 $\times$ 217,600 matrix, resulting in about 47.349 billion data points of cosine similarity scores. To do it, I will need to use following NLP packages:
1. Package:
   1. sentence_transformer
      - Using the `SentenceTransformer` module, specifically the `all-MiniLM-L6-v2` model, I convert text data into embeddings (i.e., vectorize the text).
   2. Faiss:
      1. **Vector Normalization**: To normalize the vectors to unit length (L2 norm = 1), I use `faiss.normalize_L2(embeddings_np)`.
      2. Similarity score calculation : $ \text{cosine similarity (A, B)} = \frac{A\cdot B}{||A||\ ||B||}$.
         - $||A||$ and $||B||$ are the norms of the vectors. $\Rightarrow ||A|| = ||B|| =1$
      3. Filtering Top Similar Neighbors: FAISS allows specifying `K` to retrieve only the top K most similar neighbors, which reduces computational load and focuses on the most relevant matches.

For example: N
- A = [2,3,1]
- B = [1,1,1]

L2 norms of each vector:
- $||A|| = \sqrt{2^2 + 3^2 + 1^2} = \sqrt{14}$
- $||B|| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3}$

Normalised ||A|| and ||B||
- $A_{normalised} = [\frac{2}{\sqrt{14}}, \frac{3}{\sqrt{14}}, \frac{1}{\sqrt{14}}]$
- $B_{normalised} = [\frac{1}{\sqrt{3}}, \frac{1}{\sqrt{3}}, \frac{1}{\sqrt{3}}]$

Cosine similarity: the similarity between text A and B
$$\text{cosine similarity (A, B)} = \frac{A\cdot B}{||A||\ ||B||}$$
$$\text{cosine similarity (A, B)} = \left( \frac{2}{\sqrt{14}} \times \frac{1}{\sqrt{3}}\right) + \left( \frac{3}{\sqrt{14}} \times \frac{1}{\sqrt{3}}\right) + \left( \frac{1}{\sqrt{14}} \times \frac{1}{\sqrt{3}}\right) $$

````python
# package setups for similarity test
import torch
from sentence_transformers import SentenceTransformer
import faiss  # FAISS package
# avoid warning
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module='tqdm.autonotebook')
# Hugging face log in
from huggingface_hub import login
login('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')
````

````python
# get text data ready for analysis
# I do the sort once again.
tweet_texts = df['Text_lower'].sort_values().tolist()

# Use a pre-trained BERT-based Sentence Transformer model to convert tweets to embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')

# Set up device: GPU if available, otherwise CPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = model.to(device)

# Encode tweets, use GPU if available
embeddings = model.encode(tweet_texts, convert_to_tensor=True, show_progress_bar=True, device=device)

# Convert embeddings to numpy array and normalize (required for FAISS)
embeddings_np = embeddings.cpu().numpy().astype(np.float32)
faiss.normalize_L2(embeddings_np)

# Determine whether to use GPU or CPU version of FAISS
dimension = embeddings_np.shape[1]
# if I have GPU, then we run everything using GPU. Else, we run everything using CPU
if device == 'cuda':
    # Initialize FAISS GPU resources and index
    res = faiss.StandardGpuResources()
    index = faiss.GpuIndexFlatIP(res, dimension)  # inner product search on GPU
else:
    # Use CPU FAISS index
    index = faiss.IndexFlatIP(dimension)  # inner product search on CPU

# Add embeddings to the index
index.add(embeddings_np)

# Perform a search to calculate the similarity matrix
k = 21  # Top 20 nearst neighbors. The first one is it self. So, it is always 1
sim_scores, sim_neighbor = index.search(embeddings_np, k)
# `similarity_scores` and `neighbors` contain the similarity information
````

````python
# put all high_sim items into different list for indexing purpose.
high_sim_index = []
high_sim_neighbor = []
high_sim_scores = []
sim_score_threhold = 0.9

for i in range(sim_scores.shape[0]):
  # anything above 90% cosine similarity
  # for now, I just grab all pairs where their similarity test scores are higher than 0.85
  sim_index = np.where(sim_scores[i] >= sim_score_threhold)[0]
  # size of numpy array must > 1 and the first item must not be 0
  # two conditions must be meat
  if sim_index.size == 1 and sim_index[0] == 0:
    high_sim_index.append(np.nan)
    high_sim_neighbor.append(np.nan)
    high_sim_scores.append(np.nan)
  # else, we keep the raw and matched similar texts
  else:
    high_sim_index.append(sim_index)
    high_sim_neighbor.append(sim_neighbor[i, sim_index])
    high_sim_scores.append(sim_scores[i, sim_index])

# combine everything together
# index means, the index number of tweet_text we have. I dropped rows whether there is no similarity matches.
sim_text_dict = pd.DataFrame(list(zip(high_sim_index, high_sim_neighbor, high_sim_scores)), columns = ['high_sim_index', 'high_sim_neighbor', 'high_sim_scores']).dropna()

# show me the result
sim_text_dict    # there is 58232 rows
````

Filter strategy: Take tweet_text 7 as my exmaple
1. If the user is the same
2. If the tweet post dates lie within 14 days (2 weeks) window ?
3. If the similarity is above 90%

If condition 1,2,3 are all matched, we will keep either `tweet_text[7]` or `tweet_test[156841]`, based on length of text strings. And we prefer to keep the one that come with longer text.

````python
# another method
from datetime import timedelta

# Step 1: Group similar texts into clusters
clusters = []    # This will store lists of indices that are similar to each other
visited = set()  # To track indices we've already processed
sim_score_threhold = 0.9 # remove anything that with cosine simiarity test > 0.9

for idx, (scores, neighbors) in enumerate(zip(sim_scores, sim_neighbor)):
    if idx in visited:
        continue

    # Create a new cluster for the current post
    cluster = [idx]
    visited.add(idx)

    current_user = df.iloc[idx]['User_id']
    current_date = df.iloc[idx]['Date']

    # Find similar neighbors that satisfy all conditions
    for neighbor_idx, score in zip(neighbors[1:], scores[1:]):
        if neighbor_idx in visited:
            continue

        if score >= sim_score_threhold:
            neighbor_user = df.iloc[neighbor_idx]['User_id']
            neighbor_date = df.iloc[neighbor_idx]['Date']

            # Check if they are from the same user and within 14 days
            if current_user == neighbor_user and abs((current_date - neighbor_date).days) <= 14:
                cluster.append(neighbor_idx)
                visited.add(neighbor_idx)

    clusters.append(cluster)

# Step 2: Retain only one text from each cluster (based on longest text, or first occurrence if tie)
rows_to_keep = set()

for cluster in clusters:
    # Sort by text length in descending order
    cluster_sorted_by_length = sorted(cluster, key=lambda x: df.iloc[x]['Text_length'], reverse=True)

    # Retain the first occurrence in case of ties in length
    rows_to_keep.add(cluster_sorted_by_length[0])
````
